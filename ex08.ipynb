## Q01

リスト
```
temp = [12.3, 27.9, 23.0, 21.1, 30.5, 8.9, 26.1, 28.5]
```
から、25.0以上の要素だけを取り出してリストとして出力するプログラムを作成せよ。プログラムはリストの中身が変化しても対応できるものにすること。

**サンプルケース1**
- 入力値: `なし`
- 出力値: `[27.9, 30.5, 26.1, 28.5]`

````python
temp = [12.3, 27.9, 23.0, 21.1, 30.5, 8.9, 26.1, 28.5]
mylist = []
for num in temp:
    if num >= 25:
        mylist.append(num)

print(mylist)
````

## Q02

キーボードから半角スペース区切りでいくつかの整数を入力する。そのとき、入力された整数の中で偶数のみの和を計算するプログラムを作成せよ。

例えば、キーボードから
```
1 2 3 4 5 6
```
が入力されたとき、$2 + 4 + 6$ を計算して $12$ を出力するようにせよ。

プログラムに
```
numbers = list(map(int, input().split()))
```
と書くと、キーボードから半角スペース区切りで入力された整数が、list型変数 numbers に代入される。

**サンプルケース1**
- 入力値: `1 2 3 4 5 6`
- 出力値: `12`

**サンプルケース2**
- 入力値: `5 9 15 7`
- 出力値: `0`

````python
numbers = list(map(int, input().split()))
a = 0
for i in numbers:
    if i % 2 == 0:
        a += i
print(a)
````

## Q03

リスト
```
numbers = [3, 15, 4, 32, 9, 8, 14]
```
について、奇数のみを出力し、かつ30以上の整数が出た場合には、そこでプログラムを終了させるプログラムを作成せよ。

## Q12

$N$ 本の木があり、 $i$ 番目の木には $A_i$個の木の実が実っています。シマリスは、次のルールで全ての木から木の実を収穫します。

*   実っている木の実が 10 個以下の木からは木の実を収穫しない
*   実っている木の実が 10 個より多い木からは、10 個を残して残りの全てを収穫する

シマリスが収穫する木の実の個数の合計を求めてください。

入力は以下の形式で標準入力から与えられる。

`N`

$A_1 　 A_2 　 ... 　  A_N$

​
答えを出力せよ。

**サンプルケース1**
- 入力値: 
```
3
6 17 28
```
- 出力値: `25`

````python
N = int(input())
A = list(map(int, input().split()))
S = 0
for i in range(0,N):
    if A[i] >= 10:
        S += A[i] - 10
print(S)
````
